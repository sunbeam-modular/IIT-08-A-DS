## Command Shortcuts

- esc + a : add a cell above the current cell
- esc + b : add a cell below the current cell
- enter : set focus inside the cell to enter the character and press shift + enter
- esc + dd : delete the current cell
- esc + m : convert the type of cell into markdown
- esc + y : convert the type of cell to code
- shift + enter : to execute the code
- tab : shows the intelliscence
