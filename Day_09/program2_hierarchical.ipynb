### Hierarchical Clustering

```python
# pre-requisites
import warnings
warnings.filterwarnings(action="ignore")
```

```python
# import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Mall_Customers.csv')
df.head(10)
```

```python
df.info()
```

```python
df=df.drop(['CustomerID','Gender','Age'],axis=1)
df.head()
```

```python
df.describe()
```

```python
from sklearn.cluster import AgglomerativeClustering

#create a model
hierarchical_clustering =AgglomerativeClustering(n_clusters=5)

hierarchical_clustering.fit(df)
```

```python
hierarchical_clustering.labels_
```

```python
# gives the childeren of root most class
hierarchical_clustering.children_
```

```python
clusters = hierarchical_clustering.fit_predict(df)
clusters
```

```python
plt.figure(figsize=(10,8))
colors = ['red','green','blue','black','orange']
markers =['.','^','1','s','d']

for k in range(5):
    plt.scatter(
        df['Annual Income (k$)'][clusters == k],
        df['Spending Score (1-100)'][clusters == k],
        color = colors[k],
        marker = markers[k],
        s=40, # size of the marker
        label = f"Cluster{k}"
    )
    
plt.legend()
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
```

### plotting using dendogram

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy


# find  the linked cluster
linkages = hierarchy.linkage(hierarchical_clustering.children_)

# create a dendogram
plt.figure(figsize=(15,10))
dd=dendrogram(linkages)

plt.figure(figsize=(15,10))
```

```python
# shows small linkages ..as each data point is considered as a single cluster and plotted
```
