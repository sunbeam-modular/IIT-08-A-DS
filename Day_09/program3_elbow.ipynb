### using elbow method

```python
# pre-requisites
import warnings
warnings.filterwarnings(action="ignore")
```

```python
# import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df=df.drop(['CustomerID','Gender','Age'],axis=1)
df.head()
```

```python
wss=[]
#try for 1 to 21 rows
```

```python
# import package for KMeans
from sklearn.cluster import KMeans
```

```python
for k in range(1,21):
    # create an object for Kmeans
    model = KMeans(n_clusters=k)
    
    #fit the data in object
    model.fit(df)
    
    #collect the wss
    wss.append(model.inertia_)
    # intertia = sum of squared distances of samples to their closest cluster centers
    # weighted by sample weights if provided
    
```

```python
print(wss)
```

```python
# Visualization
k_values = np.arange(1,21)
plt.plot(k_values,wss)
plt.scatter(k_values,wss)
```

```python
plt.scatter(k_values,wss)
plt.scatter(k_values[4],wss[4],color='red')
plt.xlabel("K")
plt.ylabel("WSS")
plt.title("ELBOW METHOD")
```
