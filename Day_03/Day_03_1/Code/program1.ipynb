## preprocessing

```python
# add a new column with constant value
df['dummy'] = 100

# add a column named as bonus
df['Bonus']=df['Salary']+(df['Salary']*0.10)

# add a column named total_salary
df['Total_Salary']=df['Salary']+df['Bonus']
```
