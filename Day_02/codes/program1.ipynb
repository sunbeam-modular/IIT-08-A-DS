## numpy array creation example array() and print its properties

```python
import numpy as np
```

```python
def function1():
    a1 = [10,20,30,40,50]
    print(a1)
    print(f"DType = {type(a1)}")
    print(f"a1[0]={type(a1[0])}")
    
```

```python
function1()
```

```python
def function2():
    n_array=np.array([11,22,33,44,55])
    print(n_array)
    print(f"Type={type(n_array)}")
```

```python
function2()
```

```python
def function3():
    arr=np.array([10,20,30,40,50,60])
    print(arr)

    print(f"Type = {type(arr)}")
    print(f"DType = {arr.dtype}")
    print(f"Item size= {arr.itemsize}")
    print(f"Size = {arr.size}")
    print(f"NDIM = {arr.ndim}")
    print(f"NBYTES = {arr.nbytes}")
```

```python
function3()
```

```python
def printProperties(arr):
    print(arr)
    print(f"Type = {type(arr)}")
    print(f"DType = {arr.dtype}")
    print(f"Item size= {arr.itemsize}")
    print(f"Size = {arr.size}")
    print(f"NDIM = {arr.ndim}")
    print(f"NBYTES = {arr.nbytes}")

n1 =np.array([11,22,33,44,55])
printProperties(n1)
    
    
```
