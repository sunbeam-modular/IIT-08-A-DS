```python
def function1():
    n_series=pd.Series([10,20,30,40,50,60])
    print(f"Type = {type(n_series)}")
    print(f"DType = {n_series.dtype}")
    print(f"Size = {n_series.size}")
    print(f"NDIM = {n_series.ndim}")
    print(f"NBYTES = {n_series.nbytes}")
```

## indexing

```python
s1 = pd.Series([10,20,30,40,50,60])
print(s1)
print(f"S1[2]={s1[2]}")
```

#### creating pandas series using dictionary

```python
s1=pd.Series({"Name":"Sujata","org":"Sunbeam","Course":"Python"})
print(s1)
print(f"s1[NAME]= {s1['Name']}")
print(f"s1[0]={s1[0]}")
```

```python
students=pd.Series(["s1","s2","s3","s4"])
marks=pd.Series([25,21,22,24])
print(students)
print(marks)
```

```python
marks=pd.Series([25,21,22,24])
students=pd.Series(["s1","s2","s3","s4"],index=marks)
print(students)
```

```python
def function1():
    p1 = ({"name":"P1","age":25,"id":1234})
    p2 = ({"name":"P2","age":35,"id":456})
    p3 = ({"name":"P3","age":20,"id":789})
    
    s1 = pd.Series(p1)
    s2 = pd.Series(p2)
    s3 = pd.Series(p3)
    
    df = pd.DataFrame([s1,s2,s3])

    print(s1)
    print(s2)
    print(s3)
    print(df)
    
function1()
```

```python
def function2():
    p1 = ({"name":"P1","age":25,"id":1234})
    p2 = ({"name":"P2","age":35,"id":456})
    p3 = ({"name":"P3","age":20,"marks":85})
    
    s1 = pd.Series(p1)
    s2 = pd.Series(p2)
    s3 = pd.Series(p3)
    
    df = pd.DataFrame([s1,s2,s3])
    
    print(df)
    
function2()
```
