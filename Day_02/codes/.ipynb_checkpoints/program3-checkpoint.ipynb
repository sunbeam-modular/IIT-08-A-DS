## zeros(), ones(),arange(), random()

```python
import numpy as np
```

```python
arr1=np.arange(0,10)
print(arr1)
```

```python
arr2=np.arange(10,25,2)
print(arr2)
```

```python
arr3=np.arange(1,10,2)
print(arr3)
```

```python
arr4=np.zeros(5)
print(arr4)
```

```python
arr5=np.ones(8)
print(arr5)
```

```python
n1=np.array([11,22,33,44])
n2=np.array([10,20,30,40])
print(f"N1 ={n1}")
print(f"N2 ={n2}")
print(n1+n2)
```

```python
n3=np.random.random(5)
print(n3)
```

```python
n4=np.random.randint(1,10,4)
print(n4)
```

```python
print(n1*5)   # broadcast operation
```
