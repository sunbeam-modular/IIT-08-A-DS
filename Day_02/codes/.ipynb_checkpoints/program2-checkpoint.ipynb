## Indexing,Slicing and reshape

```python
import numpy as np
```

```python
n1 = np.array([11,22,33,44,55,66,77,88])
print(n1)
```

```python
print(n1[2])       # positive indexing
```

```python
print(len(n1))
```

```python
print(n1[4])     # positive indexing
```

```python
print(n1[-2])    # negative indexing
```

```python
print(n1[-1])    # negative indexing
```

```python
print(n1[7])     # positive indexing
```

```python
print(n1[2:5])  # start[2] = slicing starts    # stop[5] = slicing ends(last index is not considered)
```

```python
print(f"NDIM={n1.ndim}")
```

```python
print(f"Shape={n1.shape}")
```

```python
n1 =n1.reshape([4,2])
print(n1)
```

```python
print(f"NDIM={n1.ndim}")
```

```python
print(f"Shape={n1.shape}")
```
